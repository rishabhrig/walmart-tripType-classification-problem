```python

from sklearn.metrics import confusion_matrix   # Model Performance Analysis, by using confusion matrix

# confusion matrix:- it is a better way to evaluate the performance of a classifier.
# The general idea is to count the number of times instances of class A are classified as class B.
# Each row in a confusion matrix represents an actual class, while each column represents a predicted class or vice versa.

#[code snippet-51]
```

```python
confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'],margins=True)

# crosstab:- This method is used to compute a simple cross-tabulation of two (or more) factors.
# By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed.
 
 # index(y_test) : array-like, Series, or list of arrays/Series, Values to group by in the rows.
 # columns(y_pred) : array-like, Series, or list of arrays/Series, Values to group by in the columns.
 # rownames : sequence, default None, If passed, must match number of row arrays passed.
 # colnames : sequence, default None, If passed, must match number of column arrays passed.
 # margins : bool, default False, Add row/column margins (subtotals).
 # Row indicates the actual values of data and columns indicate the predicted data.

 #[code snippet-52]
```

```python
from sklearn.metrics import classification_report   
                                                            
print(classification_report(y_test,y_pred)) 

# Classification Report:- As the name suggests, it is the report which explains everything about the classification.
# This is the summary of the quality of classification made by the constructed ML model

# Precision Score:- it is the accuracy of the positive predictions.
               # TP - True positive; FP - False positive
               # precision = TP/(TP+FP)                                                         
# Recall Score:- fraction of the positives correctly identified
               # FN - False Negative
               # Recall = TP/(TP+FN)
# F1 Score:- takes both precision and recall into account; 
               # F1 Score = 2(precision*recall)/(precision+recall)
# Support: It is the total entries of each class in the actual dataset. It is simply the sum of rows for every class.  

#[code snippet-53]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score,recall_score
```

```python
n_estimators=[100,300]   # Number of trees in random forest
   
max_depth=[8,16,None]     # max deapth of tree 

results_1=[]             #using custom loop for hyperparameter tuning

for i in n_estimators:
  for k in max_depth:
    model_1= RandomForestClassifier(n_estimators=i,max_depth=k,random_state=seed)
    model_1.fit(x1_train,y1_train)
    preds=model_1.predict(x1_test)
    results_1.append({'Accuracy':round(accuracy_score(y_test,preds),5),'Precision':round(precision_score(y_test,preds,average='macro'),5),'Recall':round(recall_score(y_test,preds,average='macro'),5),'F1-score':round(f1_score(y_test,preds,average='macro'),5),'n_estimators':i,'max_depth':k})
results_1=pd.DataFrame(results_1)
results_1=results_1.sort_values(by='Accuracy',ascending=False)
print(results_1)
```

```python
n_estimators=[100,300]   # Number of trees in random forest
max_features=['sqrt','log2',None]   
max_depth=[8,16,None]      

results_1=[]

for i in n_estimators:
  for j in max_depth:
    for k in max_features: 
      model_1= RandomForestClassifier(n_estimators=i,max_depth=j,max_features=k,random_state=seed)
      model_1.fit(x1_train,y1_train)
      preds=model_1.predict(x1_test)
      results_1.append({'Accuracy':round(accuracy_score(y_test,preds),5),'Precision':round(precision_score(y_test,preds,average='macro'),5),'Recall':round(recall_score(y_test,preds,average='macro'),5),'F1-score':round(f1_score(y_test,preds,average='macro'),5),'n_estimators':i,'max_depth':j,'max_features':k})
results_1=pd.DataFrame(results_1)
results_1=results_1.sort_values(by='Accuracy',ascending=False)
print(results_1)
```
